```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# 2 - Load the countries.csv file

data = pd.read_csv('countries.csv')
data.head()
```

```python
# create a dataframe for Asia

asia = data[data.continent == 'Asia']
asia.head()
```

```python
# create dataframe for Africa

africa = data[data.continent == 'Africa']
africa.head()
```

```python
# data from 1952 for Asia and Africa

asia_1952 = asia[asia.year == 1952]
africa_1952 = africa[africa.year == 1952]
```

```python
# create a histogram and subplot for 1952 data

plt.subplot(211)
plt.title('Distribution of Life Expectancy in 1952')
plt.hist(asia_1952.lifeExpectancy, range=(25, 65), edgecolor='black')
plt.ylabel('Asia')

plt.subplot(212)
plt.hist(africa_1952.lifeExpectancy, range=(25, 65), edgecolor='black')
plt.ylabel('Africa')
plt.show()
```
