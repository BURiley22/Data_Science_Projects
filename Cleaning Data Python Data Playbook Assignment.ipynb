```python
import pandas as pd
```

```python
data = pd.read_csv('Realestate.csv')
```

```python
data.head()
```

```python
data.houseAge.mean()
# mean of 'houseAge' column
```

```python
data.houseAge.min()
# lowest value
```

```python
data.houseAge.max()
# largest (or in our case oldeset) value
```

```python
data.houseAge.std()
# standard variation
```

```python
houseAge = data.houseAge

# for clarity I have created a new variable
```

```python
norm = (houseAge - houseAge.mean())/houseAge.std()
norm
# this is to normalize the data
```

```python
minmax = (houseAge - houseAge.min())/(houseAge.max() - houseAge.min())
minmax
# creates a variable called minmax to full normalize the column 
```

```python
data['houseAgeStandardized'] = minmax
# creates an additional column to the original data
```

```python
data.head()

#to verify that we have added a new column
```

```python
data.drop(columns=['numberOfConvenienceStores'], inplace=True)

# addig inplace=True will make our changes permanent
```
