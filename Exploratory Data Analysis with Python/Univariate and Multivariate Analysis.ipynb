```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
# I will start by looking at the "Carat" variable

# finding the mean

from statistics import mean

carat_mean = diamonds['carat'].mean() # pandas function
carat_mean
```

```python
# finding the mode

carat_mode = diamonds['carat'].mode() # pandas function
carat_mode
```

```python
#finding the median

carat_median = diamonds['carat'].median() # pandas function
carat_median
```

```python
# find the standard deviation

carat_std = diamonds['carat'].std() # pandas function
carat_std
```

```python
# finding the variance

carat_variance = diamonds['carat'].var() # pandas function
carat_variance
```

```python
# finding the skew

carat_skew = diamonds['carat'].skew() # pandas function
carat_skew
```

```python
# find the kurtosis

carat_kurtosis = diamonds['carat'].kurt() # pandas function
carat_kurtosis
```

```python
result = diamonds.describe()
result["carat"]
```

```python
# now we will compare the "carat" to the "price"

# Covariance

np.cov(diamonds['carat'], diamonds['price'])
```

```python
# Correlation

np.corrcoef(diamonds['carat'], diamonds['price'])
```

```python
# scatter plot

sns.scatterplot(x='price', y='carat', data=diamonds)
```

```python
# Joint Plot is not a great choice to display our data

sns.jointplot(x='price', y='carat', data=diamonds, color='#F15B2A', kind='hex')
```
